## Graphe 1
Réaliser un graphique avec deux axes. L'axe des abscisses va de 0 à 50 et s'intitule "Axe des x". L'axe des ordonnées s'intitule "Axe des y".
Tracer la droite d'équation : x = 3y. Avec x allant de 0 à 50.
Le graphe possède le titre suivant "Tracer une droite" et la droite est de couleur rouge.

```python
# Create the vectors X and Y
x = np.array(range(51))
y = 3 * x

# Create the plot and choose a color
plt.plot(x, y, color='red', label='y = 3x')

# Add a title
plt.title("Tracer une droite")

# Add X and y Label
plt.xlabel('x axis')
plt.ylabel('y axis')

# Set axis range
plt.xlim([0, 50])
plt.ylim([0, 160])

# Add a grid
plt.grid(alpha=.4,linestyle='--')

# Add a Legend
plt.legend()

# Show the plot
plt.show()
```

## Graphe 7
Nous souhaitons un graphe qui donne l'évolution du CA pour chaque catégorie (en fonction du temps).

```python
df_transactions = pd.merge(df_transactions,df_products,on='id_prod',how='left')
df_transactions = df_transactions.reindex(columns=['id_prod', 'price', 'categ', 'date', 'session_id','client_id'])
```

```python
df_transactions["month"] = df_transactions["date"].str[5:7]
df_transactions["year"] = df_transactions["date"].str[:4]
df_transactions["month_year"] = df_transactions["month"] + '-' + df_transactions["year"]
```

```python
df_transactions_v2 = df_transactions[['month_year', 'categ', 'price']].copy()
df_transactions_v2["CA"] = df_transactions_v2.groupby(['month_year', 'categ'])['price'].transform('sum')
df_transactions_v2 = df_transactions_v2.drop(columns=["price"])
df_transactions_v2 = df_transactions_v2.drop_duplicates()
```

```python
df_transactions_categ_0 = df_transactions_v2.loc[df_transactions_v2['categ'] == 0].drop(columns=["categ"])
df_transactions_categ_0['month_year'] = pd.to_datetime(df_transactions_categ_0["month_year"],format='%m-%Y',errors='coerce')
df_transactions_categ_0 = df_transactions_categ_0.sort_values('month_year')
df_transactions_categ_1 = df_transactions_v2.loc[df_transactions_v2['categ'] == 1].drop(columns=["categ"])
df_transactions_categ_1['month_year'] = pd.to_datetime(df_transactions_categ_1["month_year"],format='%m-%Y',errors='coerce')
df_transactions_categ_1 = df_transactions_categ_1.sort_values('month_year')
df_transactions_categ_2 = df_transactions_v2.loc[df_transactions_v2['categ'] == 2].drop(columns=["categ"])
df_transactions_categ_2['month_year'] = pd.to_datetime(df_transactions_categ_2["month_year"],format='%m-%Y',errors='coerce')
df_transactions_categ_2 = df_transactions_categ_2.sort_values('month_year')
df_transactions_categ_2
```

```python
plt.plot(df_transactions_categ_0["month_year"], df_transactions_categ_0["CA"], color='#d0f4de', label='Cat:0')
plt.plot(df_transactions_categ_1["month_year"], df_transactions_categ_1["CA"], color='#e4c1f9', label='Cat:1')
plt.plot(df_transactions_categ_2["month_year"], df_transactions_categ_2["CA"], color='#a9def9', label='Cat:2')

plt.title("Evolution du CA par catégorie", fontweight="bold")
plt.legend()
ax = plt.subplot(111)
ax.legend(bbox_to_anchor=(1.001, 1), ncol=1, fancybox=True, shadow=True)

plt.xlabel('mois')
plt.ylabel('profit (en $)')

ax.set_xlim([datetime.date(2021, 3, 1), datetime.date(2022, 2, 1)])

plt.grid(alpha=.4,linestyle='--')

months = ['03/2021', '04/2021', '05/2021', '06/2021', '07/2021', '08/2021', '09/2021', '10/2021', '11/2021', '12/2021', '01/2022', '02/2022']
plt.xticks(df_transactions_categ_0["month_year"], months, rotation = 45)

plt.show()
```

## Graphe 9
Nous souhaitons un graphe qui donne la répartition du CA, pour chaque mois, en fonction du genre du client.

```python
df_clients = pd.read_csv("df_clients_clean.csv", sep=",")
df_clients
```

```python
df_clients["month"] = df_clients["date"].str[5:7]
df_clients["year"] = df_clients["date"].str[:4]
df_clients["month_year"] = df_clients["month"] + '-' + df_clients["year"]
df_clients['month_year'] = pd.to_datetime(df_clients["month_year"],format='%m-%Y',errors='coerce')
df_clients
```

```python
df_clients_v2 = pd.merge(df_clients,df_customers,on='client_id',how='left')
df_clients_v3 = df_clients_v2.reindex(columns=['sex', 'price', 'month_year'])
```

```python
df_clients_v3["CA"] = df_clients_v3.groupby(['month_year', 'sex'])['price'].transform('sum')
df_clients_v3 = df_clients_v3.drop(columns=["price"])
df_clients_v3 = df_clients_v3.drop_duplicates()
df_clients_v3
```

```python
df_clients_femmes = df_clients_v3.loc[df_clients_v3['sex'] == 'f'].rename(columns={"sex" : "femmes"})
df_clients_hommes = df_clients_v3.loc[df_clients_v3['sex'] == 'm'].rename(columns={"sex" : "hommes"})

df_clients_v4 = pd.DataFrame({'month_year_f': df_clients_femmes['month_year'], 'month_year_m': df_clients_hommes['month_year'], 'CA_femmes': df_clients_femmes['CA'], 'CA_hommes': df_clients_hommes['CA']})
df_clients_v4
```

```python
df_clients_v4['month_year_both'] = df_clients_v4['month_year_f'].combine_first(df_clients_v4['month_year_m'])
df_clients_v4 = df_clients_v4.sort_values('month_year_both').drop(columns=["month_year_f", "month_year_m"])
df_clients_v4
```

```python
df_clients_v5 = df_clients_v4.groupby(df_clients_v4['month_year_both']).aggregate({'CA_femmes': 'sum', 'CA_hommes': 'sum'})
df_clients_v5 = df_clients_v5.reset_index(inplace = False)
df_clients_v5
```

```python
df_clients_v5.plot.bar(x = 'month_year_both', y = ['CA_femmes', 'CA_hommes'])

plt.show()

# à mettre en forme
```
